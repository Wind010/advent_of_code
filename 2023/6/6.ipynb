```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def calculate_distance(time, speed):
    return (time - speed) * speed

speed = np.linspace(0, 7, 50)
travel = calculate_distance(7, speed)
distance_limit = 9
plt.plot(speed, travel)
plt.scatter(range(0, 8), calculate_distance(7, np.arange(0, 8)))
plt.axhline(distance_limit, color='red')
```
